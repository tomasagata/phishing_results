```python
import pandas as pd
import numpy as np

df = pd.read_csv("SSN Project_October 14, 2024_03.25.csv")

df = df.drop(axis=1, labels=[
  "startDate",
  "endDate",
  "status",
  "progress",
  "duration",
  "finished",
  "recordedDate",
  "responseId",
  "distributionChannel",
  "userLanguage"
])

df = df - 1

df
```

```python
is_phishing = {
  "E1.1": "Phishing",
  "E2.1": "Phishing",
  "E3.1": "Phishing",
  "E4.1": "Phishing",
  "E5.1": "Phishing",
  "E6.1": "Phishing",
  "E7.1": "Phishing",
  "E8.1": "Phishing",
  "E9.1": "Phishing",
  "E10.1": "Phishing",
  "E11.1": "Phishing",
  "E12.1": "Phishing",
  "E13.1": "Phishing",
  "E14.1": "Phishing",
  "E15.1": "Legit",
  "E16.1": "Legit",
  "E17.1": "Legit",
  "E18.1": "Legit",
  "E19.1": "Legit",
  "E20.1": "Legit",
  "E21.1": "Legit",
}

# Replaces 0.0 with "Legit" and 1.0 with "Phishing"
df.replace(to_replace=dict(zip(is_phishing.keys(), [0.0] * 21)), value="Legit", inplace=True)
df.replace(to_replace=dict(zip(is_phishing.keys(), [1.0] * 21)), value="Phishing", inplace=True)

df
```

```python
column_groups = {
  "prediction": {
    "control":       ["E8.1", "E10.1", "E14.1", "E15.1", "E18.1", "E21.1"],
    "no_subject":    ["E4.1", "E7.1", "E9.1"  ],
    "no_sender":     ["E2.1", "E13.1", "E16.1"],
    "no_images":     ["E3.1", "E19.1", "E20.1"],
    "no_signature":  ["E1.1", "E6.1", "E12.1" ],
    "no_targeted":   ["E5.1", "E11.1", "E17.1"],
  },
  "confidence": {
    "control":      ["E8.2", "E10.2", "E14.2", "E15.2", "E18.2", "E21.2"],
    "no_subject":   ["E4.2", "E7.2", "E9.2"  ],
    "no_sender":    ["E2.2", "E13.2", "E16.2"],
    "no_images":    ["E3.2", "E19.2", "E20.2"],
    "no_signature": ["E1.2", "E6.2", "E12.2" ],
    "no_targeted":  ["E5.2", "E11.2", "E17.2"]
  }
}
```

```python
results = {
  "control_hit":       [],
  "no_subject_hit":    [],
  "no_sender_hit":     [],
  "no_images_hit":     [],
  "no_signature_hit":  [],
  "no_targeted_hit":   [],
  "control_subj_conf":      [],
  "no_subject_subj_conf":   [],
  "no_sender_subj_conf":    [],
  "no_images_subj_conf":    [],
  "no_signature_subj_conf": [],
  "no_targeted_subj_conf":  []
}
```

```python
# Get values per row
for row in range(len(df.index)):

  # List prediction results per group
  for group, questions in column_groups["prediction"].items():
    for question in questions:
      response = df.loc[row, question]

      if (type(response) == float): # Quick hack to check for no-responses
        continue

      answer = is_phishing[question]

      if (response == answer): 
        results[(group+"_hit")].append(1)
        continue
      results[(group+"_hit")].append(0)

  # Get averages of confidence values per group
  for group, questions in column_groups["confidence"].items():
    results[(group+"_subj_conf")].append(
      df.loc[row, questions].squeeze().dropna().mean()
    )
```

```python
import scipy.stats as st

result_data = {
  "mean": [],
  "std_err_norm": [],
  "conf_intvl_norm_95_lower_edge": [],
  "conf_intvl_norm_95_higher_edge": [],
  "conf_intvl_norm_95_plus_minus": [],
  # "std_deviation_t": [],
  # "conf_intvl_t_95_lower_edge": [],
  # "conf_intvl_t_95_higher_edge": [],
  # "conf_intvl_t_95_plus_minus": []
}


for group, values in results.items():
  mean = np.mean(values)
  std_err_norm = st.sem(values, 
                ddof=0)
  conf_intvl_norm = st.norm.interval(confidence=0.95,
                loc=mean,
                scale=std_err_norm)
  
  result_data["mean"].append(mean)
  result_data["std_err_norm"].append(std_err_norm)
  result_data["conf_intvl_norm_95_lower_edge"].append(conf_intvl_norm[0])
  result_data["conf_intvl_norm_95_higher_edge"].append(conf_intvl_norm[1])
  result_data["conf_intvl_norm_95_plus_minus"].append((conf_intvl_norm[1] - conf_intvl_norm[0])/2)

  # std_err_t = st.sem(values, 
  #               ddof=1)
  # conf_intvl_t = st.t.interval(confidence=0.95,
  #               df=len(values)-1,
  #               loc=np.mean(values),
  #               scale=st.sem(values))
  

  # result_data["std_deviation_t"].append(std_err_t)
  # result_data["conf_intvl_t_95_lower_edge"].append(conf_intvl_t[0])
  # result_data["conf_intvl_t_95_higher_edge"].append(conf_intvl_t[1])
  # result_data["conf_intvl_t_95_plus_minus"].append((conf_intvl_t[1] - conf_intvl_t[0])/2)

final = pd.DataFrame(data=result_data, index=results.keys())
final.to_csv("results.csv")

final
```

```python
import matplotlib.pyplot as plt

x = np.arange(0,100,0.1)
control_mean = final.loc["control_hit", "mean"] * 100
control_std_err = final.loc["control_hit", "std_err_norm"] * 100
no_subj_mean = final.loc["no_subject_hit", "mean"] * 100
no_subj_std_err = final.loc["no_subject_hit", "std_err_norm"] * 100

fig, ax = plt.subplots()
ax.grid(linestyle="-", linewidth=0.3)
ax.set_title("Figure 1")
ax.set_xlabel("Average probability of hit")

ax.plot(x, st.norm.pdf(x, control_mean, control_std_err), linestyle="dashed", label="control")
ax.plot(x, st.norm.pdf(x, no_subj_mean, no_subj_std_err), label="No subject")
ax.legend()
plt.show()
```
