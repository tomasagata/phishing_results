```python
import pandas as pd
import scipy.stats as st
import numpy as np

df = pd.read_csv("results.csv", index_col=0)

df
```

```python
import matplotlib.pyplot as plt

x = np.arange(0,5,0.01)
control_mean = df.loc["control_subj_conf", "mean"]
control_std_err = df.loc["control_subj_conf", "std_err_norm"]
no_subj_mean = df.loc["no_subject_subj_conf", "mean"]
no_subj_std_err = df.loc["no_subject_subj_conf", "std_err_norm"]

fig, ax = plt.subplots()
ax.grid(linestyle="-", linewidth=0.3)
ax.set_title("Figure 6")
ax.set_xlabel("Subject's confidence in answer")

ax.plot(x, st.norm.pdf(x, control_mean, control_std_err), linestyle="dashed", label="control")
ax.plot(x, st.norm.pdf(x, no_subj_mean, no_subj_std_err), label="No subject")
ax.legend()
plt.show()
```
